#### import Packgaes

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

#### Define Vector

```python
# define vector with numpy array
r = np.array([1, 1])
```

#### Plot vector in coordinate system

```python
# plot vector using red dashed line :       # plot([x_1, x_2], [y_1, y_2])
plt.plot([0, r[0]], [0, r[1]], "r--")       # a line goes from [0, 0] to [r[0], r[1]]

# draw endpoint of vector using a dot
plt.plot(r[0], r[1], marker="o")

# customizing figure
plt.grid(linestyle="--")
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# accentuate the origin
ca = plt.gca()                              # get the current Axes
plt.plot(ca.get_xlim(), [0, 0], color="k")
plt.plot([0, 0], ca.get_ylim(), color="k")
plt.show()
```
