#### Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Define vector and scalars

```python
# vector
v = np.array([3, 4])

# list of scalars
a = [2, 0.5, -1, 3, -0.7, -3]
```

```python
# plot the vector v
plt.plot([0, v[0]], [0, v[1]],
         marker="v", ms=5, color="purple", lw=3)

# plot the scaled vector
for i in range(len(a)):
    a_v = a[i] * v  #  scaled vector
    plt.plot([0, a_v[0]], [0, a_v[1]], 'ob--', lw=0.5, ms=4)
    plt.text(a_v[0]+0.5, a_v[1], a[i])

plt.grid()         
plt.xlabel('x')
plt.ylabel('y')
xlim = max(abs(v[0]), abs(a_v[0])) * 1.3
ylim = max(abs(v[1]), abs(a_v[1])) * 1.3
plt.axis((-xlim, xlim, -ylim, ylim))
plt.show()
```
